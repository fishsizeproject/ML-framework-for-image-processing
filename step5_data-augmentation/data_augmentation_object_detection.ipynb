### Import the required libraries

```python
import pandas as pd
```

## Apply vertical flip to annotations file

```python
# read annotations file
 
annotations_df = pd.read_csv(r'/content/deeper_data/annotated_images_pilot_study/annotations_for_model_6spp_good_qual.csv',header=None)
```

```python
annotations_df = annotations_df.sort_values(by=[1])
annotations_df.head()
```

```python
len(annotations_df)
```

```python
xmin = annotations_df[3]
ymin = 1-annotations_df[8]
xmax = annotations_df[7]
ymax = 1-annotations_df[4]
filename = 'vertical-flip_' + annotations_df[1].astype(str)

annotations_vertical_flip = annotations_df

annotations_vertical_flip['x_min'] = xmin
annotations_vertical_flip['y_min'] = ymin
annotations_vertical_flip['x_max'] = xmax
annotations_vertical_flip['y_max'] = ymax
annotations_vertical_flip['filename'] = filename
annotations_vertical_flip.head(5)
```

```python
annotations_vertical_flip.columns = ['dataset','filename_old','species', 'x_min', 'y_min', 'empty1','empty2', 'x_max', 'y_max',
                                       'xmin_vertical_flip','ymin_vertical_flip','xmax_vertical_flip','ymax_vertical_flip', 'filename']
```

```python
annotations_vertical_flip = annotations_vertical_flip[['dataset', 'filename','species', 'xmin_vertical_flip','ymin_vertical_flip','empty1','empty2',
                                                           'xmax_vertical_flip','ymax_vertical_flip', 'empty1','empty2',]]
annotations_vertical_flip.head(5)
```

```python
annotations_vertical_flip.to_csv('annotations_vertical_flip.csv')
```
